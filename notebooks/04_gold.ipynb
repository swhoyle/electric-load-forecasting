```python
import os
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option("display.max_columns", 200)
pd.set_option("display.width", 160)
```

```python
INTERVAL = "5m"  # "1m", "5m", "10m"

paths = {
    "1m":  f"../data/gold/power_load_1m_all_features.parquet",
    "5m":  f"../data/gold/power_load_5m_all_features.parquet",
    "10m": f"../data/gold/power_load_10m_all_features.parquet",
}

path = paths[INTERVAL]

df = pd.read_parquet(path)
df["timestamp"] = pd.to_datetime(df["timestamp"])
df = df.sort_values("timestamp").reset_index(drop=True)

print("Loaded:", path)
print("Shape:", df.shape)
display(df.head())
display(df.tail())
```
