Explore the raw data and structure

```python
import pandas as pd
import numpy as np
from scipy.io import loadmat

RAW_PATH = '../data/raw/P_data.mat'
```

```python
raw = loadmat(RAW_PATH)

print("Keys in raw data:")
for key in raw.keys():
    print(key)
```

```python
p_data = raw['P_data']
print("P_data")
print("shape:", p_data.shape)
print("type:", p_data.dtype)
print("First 3 records:")
print(p_data[:3])
```

```python
day_data = raw["day_data"]
print("day_data")
print("shape:", day_data.shape)
print("type:", type(day_data))
print("Data:")
print(day_data)
```

```python
day_class = raw["day_class"]
print("day_class")
print("shape:", day_class.shape)
print("type:", type(day_class))
print("Data:")
print(day_class)
```
